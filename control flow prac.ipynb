```python

#Write a program that calculates the sum of the digits of a number input by the user using a while loop.
# Get input from the user
num = int(input("Enter an integer: "))

# Make a copy of the number for the final print statement
temp = num
digit_sum = 0

# Convert negative numbers to positive to handle them correctly
if temp < 0:
    temp = abs(temp)

# Process each digit using a while loop
while temp > 0:
    digit = temp % 10       # Get the last digit
    digit_sum += digit      # Add the digit to the running total
    temp = temp // 10       # Remove the last digit from the number

print(f"The sum of the digits of {num} is: {digit_sum}")
```

```python
#Write a program that calculates the factorial of a number input by the user using a while loop.
number = int(input("enter a number"))
original_number = number
factorial = 1
while number > 0:
    factorial = factorial * number
    number = number - 1  # Subtract 1 to eventually end the loop

print(f"The factorial of {original_number} is {factorial}")
```
