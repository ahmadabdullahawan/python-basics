```python
#data aggregation and grouping 
df.groupby('Product Category')['Total Revenue'].mean()
```

```python
#based on two categories 
df.groupby(['Product Category','Region'])['Total Revenue'].sum()
```

```python
#multiple aggregation functions
df.groupby(['Product Category','Region'])['Total Revenue'].agg(['count','sum','mean'])
```

```python
#mering and joining 
data1={"key":["A","B","D"],"Value1":[1,2,3]}
data2={"key":["A","B","E"],"Value2":[1,5,4]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
print(df1)
```

```python
print(df2)
```

```python
pd.merge(df1,df2,on='key',how="inner")
```

```python
pd.merge(df1,df2,on='key',how="outer")
```

```python
pd.merge(df1,df2,on='key',how="left")
```

```python
pd.merge(df1,df2,on='key',how="right")
```
